### Matplotlib Tutorial

##### 1. how to import required libraries

```python
# import required libs
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('Big_Japan_vs_China_Technology.csv')
```

##### 2. how to generate some random data

```python
X_data = df['R&D Investment (in USD)']  #生成1000个随机数
Y_data = df['Tech Exports (in USD)']
```

```python
print(X_data)   
```

```python
print(Y_data)
```

##### 3. how to make a scatter plot?

```python
plt.scatter(X_data, Y_data, c = "#00f", s = 150, marker= "*", alpha = 0.3)

plt.show()
```

##### 4. how to make a line plot?

```python
# Generate some data
years = [2006 + x for x in range (16)]
weights = [80, 83, 84, 85, 86, 82, 81, 82,
           79, 83, 80,82, 82, 83, 81, 79]
```

```python
# By default plt.plot will make a line plot
plt.plot(years, weights, c = "g", lw = 2, linestyle = "--")

plt.show()
```

##### 5. how to make a bar plot?

```python
x = ["C++", "C#", "Python", "Java", "Go"]
y = [20, 50, 140, 8 , 45]

plt.bar(x, y , color = "r")
plt.show()
```

```python
# Using additional option in the plt.bar function
plt.bar(x, y , color ="r", align ="edge", width=0.5, edgecolor="green", lw=6)

plt.show()
```

##### 6. how to make a histogram?

```python
ages = np.random.normal(20, 1.5, 1000)

plt.hist(ages)
plt.show()
```

```python
plt.hist(ages, bins = 20)
plt.show()
```

```python
plt.hist(ages, bins = 20, cumulative=True)
plt.show()
```

##### 7. how to make a pie chart?

```python
langs = ["C++", "C#", "Python", "Java", "Go"]
votes = [50, 24, 14, 6, 17]

plt.pie(votes, labels=langs)
plt.show()
```

```python
explodes = [0.3, 0, 0 ,0, 0]

plt.pie(votes, labels=langs, explode=explodes)
plt.show()
```

```python
explodes = [0, 0, 0 ,0, 0]

plt.pie(votes, labels=langs, explode=explodes, autopct="%0.2f%%", pctdistance=1.3)
plt.show()
```

##### 8. how to make a box plot?

```python
heights = np.random.normal(172, 8, 300)

plt.boxplot(heights)
plt.show()
```

##### 9. how to add more features in a chart?

```python
years = [2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021]
income = [55, 56, 62, 61, 72, 72, 73, 75]

plt.plot(years, income)
plt.show()
```

##### 11. how to add a legend to the plot?

```python
stock_a = [100, 102, 105, 99, 101, 100, 103]
stock_b = [97, 98, 95, 100, 102, 103, 99]
stock_c = [110, 112, 115, 111, 108, 107, 109]

plt.plot(stock_a, label = "Company A")
plt.plot(stock_b, label = "Company B")
plt.plot(stock_c, label = "Company C")

plt.legend()

plt.show()
```

##### 12. how to use different styles in matplotlib?

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style

style.use("ggplot")
```

```python
votes = [10, 12, 15, 18, 20, 2]
people = ["A", "B", "C", "D", "E", "F"]

plt.pie(votes, labels=None)
plt.legend(labels=people)

plt.show()
```

```python
# Use of dark background style
style.use("dark_background")

votes = [10, 12, 15, 18, 20, 2]
people = ["A", "B", "C", "D", "E", "F"]

plt.pie(votes, labels=None)
plt.legend(labels=people)

plt.show()
```

##### 13. how to plot multiple graphs?

```python
x1 = np.random.random(100)
y1 = np.random.random(100)  

x2 = np.arange(100)
y2 = np.random.random(100)
```

```python
style.use("ggplot")
plt.figure(1)
plt.scatter(x1, y1)

plt.figure(2)
plt.plot(x2, y2)

plt.show()
```

```python
x = np.arange(100)

fig, axs = plt.subplots(2, 2)

axs[0, 0].plot(x, np.sin(x))

axs[0, 1].plot(x, np.cos(x))

axs[1, 0].plot(x, np.random.random(100))

axs[1, 1].plot(x, np.tan(x))

plt.show()
```

##### 14. how to make 3D plot in matplotlib?

```python
ax = plt.axes(projection = "3d")

x = np.random.random(100)
y = np.random.random(100)
z = np.random.random(100)

ax.scatter(x, y, z)
ax.set_title("3D Plot")
ax.set_xlabel("test")

plt.show()
```

```python
ax = plt.axes(projection = "3d")

x = np.arange(100)
y = np.sin(x)
z = np.cos(x)

ax.scatter(x, y, z)
ax.set_title("3D Plot")
ax.set_xlabel("test")

plt.show()
```

##### 15. how to make animated plots in matplotlib?

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style

import random
```

```python
heads_tails = [0, 0]
```

```python
for _ in range((10)):
    heads_tails[random.randint(0, 1)] +=1
    plt.bar(["Heads", "Tails"], heads_tails, color=['red', 'blue'])
    plt.pause(0.001)

plt.show()
```
