# Data Cleaning in Pandas

```python
import pandas as pd
```

```python
df = pd.read_excel(r"cust_call_list.xlsx")
df
```

```python
df = df.drop_duplicates() #删除重复行
df
```

```python
df = df.drop(columns = "Not_Useful_Column")  #删除列
df
```

```python
#df["Last_Name"] = df["Last_Name"].str.lstrip("...")
#df["Last_Name"] = df["Last_Name"].str.lstrip("/")
#df["Last_Name"] = df["Last_Name"].str.rstrip("_")
df["Last_Name"] = df["Last_Name"].str.strip("123._/")   #删除last name列中开头和结尾的123，/

df
```

```python
#df["Phone_Number"] = df["Phone_Number"].str.replace('[^a-zA-Z0-9]','')

#df["Phone_Number"].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])

#df["Phone_Number"] = df["Phone_Number"].apply(lambda x: str(x))

#df["Phone_Number"] = df["Phone_Number"].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])

#df["Phone_Number"] = df["Phone_Number"].str.replace('NaN','')   #无效，因为NaN是特殊值，不是字符串

df["Phone_Number"] = df["Phone_Number"].str.replace('N/a','')

df["Phone_Number"] = df["Phone_Number"].str.replace('\\',"-")
df["Phone_Number"] = df["Phone_Number"].str.replace('/','-')
df["Phone_Number"] = df["Phone_Number"].str.replace('|','-')
df
```

```python
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('Yes','Y')

df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('No','N')
df["Paying Customer"] = df["Do_Not_Contact"].str.replace('No','N')
df["Paying Customer"] = df["Do_Not_Contact"].str.replace('No','N')
df
```

```python
df = df.replace('N/a','')
df = df.replace('NaN','')


df=df.fillna('') #删除所有NaN
df
```

```python
for x in df.index:
    if df.loc[x, "Do_Not_Contact"] == 'Y':
        df.drop(x, inplace=True)
#删除所有不愿联系的客户信息
df
```

```python
for x in df.index:
    if df.loc[x, "Phone_Number"] == '':
        df.drop(x, inplace=True)

df

#Another way to drop null values
#df = df.dropna(subset="Phone_Number"), inplace=True)
```

```python
df = df.reset_index(drop=True)
df
```
