# Merge, Join, and Concatenate 

```python
import pandas as pd
```

```python
df1 = pd.read_csv("LOTR.csv")
df1
```

```python
df2 = pd.read_csv("LOTR2.csv")
df2
```

```python
df1.merge(df2, how = 'inner', on = ['FellowshipID', 'FirstName'])
#how='inner': 指定合并方式为内连接。内连接会保留两个 DataFrame 中在指定列上匹配的行。如果某一行在 df1 或 df2 中找不到匹配项，该行将被舍弃。
```

```python
df1.merge(df2, how = 'outer')
#how='outer': 指定合并方式为外连接。外连接会保留 df1 和 df2 中的所有行。对于那些在连接列上没有匹配的行，结果中的其他列会填充 NaN（表示缺失值）
```

```python
df1.merge(df2, how = 'left')
#包含 df1 中的所有行以及 df2 中在指定列上匹配的行
```

```python
df1.merge(df2, how = 'right')
#包含 df2 中的所有行以及 df1 中在指定列上匹配的行
```

```python
df1.merge(df2, how = 'cross')
#笛卡尔积
```

```python
df1.join(df2, on = 'FellowshipID', how = 'outer', lsuffix = '_Left',rsuffix = '_Right')
#如果 df1 和 df2 有相同的列名，在 df1 中的这些列将加上 _Left 后缀以区分。
```

```python
df4 = df1.set_index('FellowshipID').join(df2.set_index('FellowshipID'), lsuffix = '_Left',rsuffix = '_Right', how = 'outer')
#把FellowshipID作为索引之后排序
df4
```

```python
pd.concat([df1,df2], join = 'outer', axis = 1)
#向右合并（axis=1）两个数据
```
